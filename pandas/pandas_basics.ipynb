## Indexing


### iloc : integer location

```
Iloc uses standard python indexing so [0:10] will select entries from 0,...,9
```



````python
data = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]},index=['X','Y','Z'])

selected_data = data.iloc[0:3, 0:2]
print(selected_data)
````

### loc : select by dataframe label
```
Loc indexing different than standard python : [0,10] will select 0,...,10
```

````python
selected_data = data.loc['X':'Z', 'A':'B']
print(selected_data)
````
